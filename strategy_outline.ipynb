# Overall Goal

The desire here is to combine my newfound coding ability with a simple idea: **invest for long** and **invest at low**. 

What do I mean by this? 

**Investing long:** this means two things: (1) only taking long positions in a company and (2) holding that position for a minimum of 40 years. 

**Investing low:** the market is cyclical. Therefore, stocks prices will inevitably drop here and there. Therefore, only invest in campanies when their stock price has dropped a certain amount (will require backtesting to obtain this figure)

# Criteria

## How will companies be chosen? 

Companies will be chosen based off of the 40-year rule. That rule, as defined by me, is that I will only invest in companies that will still be in existence 40 years from now (2061). 

This list of companies will be generated by looking into companies of history that have remained highly profitable for a minimum of 40 years. 

## How will I know when to buy

This figure will also need to be calculated through backtesting. Once obtained, the trading bot will buy more of the stock every time that the price drops by a certain calculated percentage. Once the stock is purchased, the buyer will reset to reflect that point of purchase as the most recent high.  

## How much of my liquidity should I allocate to each company and to each time a position is deepened?

Because the intention is to continually be adding liquidity to the portfolio, there will be a need to appropriately allocate these funds. Through testing, I will figure out the percentage of the liquidity to allocate to each position. This will likely be done through trial and error. 

# Steps to Completion

## Step 1: Filter for Best Companies

The algorithm should only reevaluate on a yearly basis. If the companies still fall into the top 50, they will remain. If a company drops out of the top 50, the position in that company will be liquidated and reallocated to the top company on the list that is not already in rotation. 

## Step 2: Find Percentage for Buy

Figure out what percentage to buy in. This will be done through backtesting with maximum return in mind. 

## Step 3: Develop Allocation Strategy

FIgure out how much to allocate to a position once a buying opportunity is presented. This will also be done with maximum return in mind.

## Step 4: Figure out a Failsafe

What happens if the company that the algorithm would like buy into is dropping fast? Do I want to algorithm to buy on every percentage drop if that percentage drop happens multiple times in a single day? This will also have to be tested to figure out what the best and safest configuration is for a possibility such as this.
