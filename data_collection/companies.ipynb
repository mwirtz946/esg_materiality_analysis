# Imports

```python
# All necessary imports
from bs4 import BeautifulSoup
import requests

import pickle

import warnings
warnings.simplefilter("ignore")
```

# Getting Companies

```python
# Getting site
page = requests.get('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
soup = BeautifulSoup(page.content, 'html.parser')

# Finding table
table = soup.find(id='constituents').tbody

# Getting table rows
table_rows = table.find_all('tr')

# Finding company in row
companies = []
for row in table_rows[1:510]:
    elements = row.find_all('td')
    company = elements[1].text
    companies.append(company)

# Finding ticker in row 
tickers = []
for row in table_rows[1:510]:
    elements = row.find_all('td')
    ticker = elements[0].text
    ticker = ticker.replace('\n','')
    tickers.append(ticker)

# Merging companies and tickers into tuples
companies_n_tickers = [(companies[i], tickers[i]) for i in range(0, len(tickers))]

# Removing class and series for ease of use in data scraping
problem_tuples = [('Alphabet Inc. (Class A)', 'GOOGL'),
                  ('Alphabet Inc. (Class C)', 'GOOG'),
                  ('Discovery, Inc. (Series A)', 'DISCA'),
                  ('Discovery, Inc. (Series C)', 'DISCK'),
                  ('Fox Corporation (Class A)', 'FOXA'),
                  ('Fox Corporation (Class B)', 'FOX'),
                  ('News Corp (Class A)', 'NWSA'),
                  ('News Corp (Class B)', 'NWS'),
                  ('Under Armour (Class A)', 'UAA'),
                  ('Under Armour (Class C)', 'UA')]

# Finalizing list by removing unwanted items
companies_n_tickers = [item for item in companies_n_tickers if item not in problem_tuples]
```

```python
# Writing tuple list to pickle file
with open('companies_n_tickers.pickle','wb') as f:
    pickle.dump(companies_n_tickers, f)
```
